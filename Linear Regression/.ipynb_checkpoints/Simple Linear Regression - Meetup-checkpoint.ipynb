```python
# Import the libraries
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('data.csv',header=None)
x = df[0]
y = df[1]
```

```python
# Plot the data
plt.title("Simple plot")
plt.xlabel("X - values")
plt.ylabel("Y - values")
plt.plot(x,y,'o')
plt.show()
```

```python
# Find the mean and plot it along
mean_y = y.mean()
print(mean_y)
mean_y_arr = [int(mean_y)]*len(y)
plt.title("Simple plot with mean")
plt.xlabel("X-values")
plt.ylabel("Y-values")
plt.plot(x,y,'o',x,mean_y_arr,'r--')
plt.show()
```

```python
# Perfrom linear regression
from numpy import array
regressor = LinearRegression()
regressor.fit(array(x).reshape(-1,1),array(y).reshape(-1,1))
regressor.predict(array(x).reshape(-1,1)).reshape(1,-1)
print(regressor.coef_)
print(regressor.intercept_)
plt.title("Simple linear regression plot")
plt.xlabel("X-values")
plt.ylabel("Y-values")
plt.plot(x,y,'o',x,regressor.predict(array(x).reshape(-1,1))
,'--')
plt.show()
```

```python
def compute_error_for_given_points(b, m, points):
    total_error = 0
    for i in range(0,len(points)):
        x = points[i, 0]
        y = points[i, 1]
        total_error += (y - (m * x + b)) ** 2
    return total_error/float(len(points))
```

```python
print('\tError computation: ')
points = df.as_matrix()
print(compute_error_for_given_points(regressor.coef_, regressor.intercept_,points))
```
